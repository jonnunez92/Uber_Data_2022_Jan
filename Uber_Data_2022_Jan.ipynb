# Clean Data

## Duplicates and Missing Data

```python
df.isna().sum()
```

## Convert to datetime dtype

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

# Zone Lookup

```python
df_zone = pd.read_csv('Data/taxi+_zone_lookup.csv')
df_zone.head()
```

## Pickup Borough Dict

```python
df_zone['Borough'][df_zone['LocationID'] == 1].item()
```

```python
borough_dict = {}

for loc in df_zone['LocationID']:
    borough_dict[loc] = df_zone['Borough'][df_zone['LocationID'] == loc].item()
```

```python
borough_dict
```

```python
df['PUBorough'] = df['PULocationID'].map(borough_dict)
df.head(3)
```

## Pickup Zone Dict

```python
zone_dict = {}

for loc in df_zone['LocationID']:
    zone_dict[loc] = df_zone['Zone'][df_zone['LocationID'] == loc].item()
```

```python
zone_dict
```

```python
df['PUZone'] = df['PULocationID'].map(zone_dict)
df.head(3)
```

```python
df.columns
```

```python
df = df[['VendorID', 'tpep_pickup_datetime', 'tpep_dropoff_datetime',
       'passenger_count', 'trip_distance', 'RatecodeID', 'store_and_fwd_flag',
       'PULocationID', 'PUBorough', 'PUZone', 'DOLocationID', 'payment_type', 'fare_amount', 'extra',
       'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
       'total_amount', 'congestion_surcharge', 'airport_fee']]
```
